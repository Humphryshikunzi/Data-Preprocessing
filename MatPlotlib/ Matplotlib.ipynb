###  Ploting more than one line graph on same plot

```python
fig, a =  plt.subplots(2,2, figsize=(15,15))

x_values = np.arange(0,9)

Temp_1 = [23, 33, 40, 46, 52, 64, 69, 73, 79]
Sensor_1 = [20, 30, 37, 43, 49, 61, 66, 70, 76]

Temp_2 = [23, 30, 38, 45, 51, 65, 70, 75, 81]
Sensor_2 =[20, 27, 35, 42, 47, 62, 67, 72, 78]

Temp_3 = [22, 27, 37, 43, 50, 62, 69, 74, 79]
Sensor_3 = [19, 24, 34, 40, 50, 59, 66, 71, 76]

l1 , = a[0][0].plot(x_values, Temp_1,  label = "Temp_1",  color='r')
l2 , = a[0][1].plot(x_values, Temp_2,  color='b')
l3,  = a[1][0].plot(x_values, Temp_3,  color='g')
l4 , = a[0][0].plot(x_values, Sensor_1, color='y' )
l5,  = a[0][1].plot(x_values, Sensor_2,  color='k')
l6 ,  = a[1][0].plot(x_values, Sensor_3,  color='m')

plt.legend([l1, l2, l3, l4, l5, l6], ["Temp_1", "Temp_2", "Temp_3", "Sensor_1", "Sensor_2", "Sensor_3" ])

a[0][0].set_title('Graph of Temp and Sensor values')
a[0][0].set_xlabel('Units')
a[0][0].set_ylabel('Temp, Degrees Celcius')



a[0][1].set_title('Graph of Temp and Sensor values')
a[0][1].set_xlabel('Units')
a[0][1].set_ylabel('Temp, Degrees Celcius')




a[1][0].set_title('Graph of Temp and Sensor values')
a[1][0].set_xlabel('Units')
a[1][0].set_ylabel('Temp, Degrees Celcius')



plt.show()
```

### Subplots

```python
# data to plot

n_groups = 7
Standard_Solution = (0,20,50,100,200,1000,2000)
Turbidimeter_Value=(1.1,20.3,49.1,102.3,205.2,1008.7,2012.8)
Probe_1_Sensor_Value=(0.9,20.2,48.8,101.5,202.5,1000,1000)
Probe_2_Sensor_Value=(0.8,21.1,52.1,98.1,201.5,1000,1000)
Probe_3_Sensor_Value=(1.3,19.8,50.6,106.5,205.1,998.9,1000)


# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.1
opacity = 0.8

rects1 = plt.bar(index, Standard_Solution, bar_width,alpha=opacity,color='r',label='Standard Solution')

rects2 = plt.bar(index + bar_width, Turbidimeter_Value, bar_width,alpha=opacity,color='b',label='Turbidimeter_Value')

rects3 = plt.bar(index + (2*(bar_width)), Probe_1_Sensor_Value, bar_width,alpha=opacity,color='g',label='Probe_1_Sensor_Value')

rects4 = plt.bar(index + (3*(bar_width)), Probe_2_Sensor_Value, bar_width,alpha=opacity,color='m',label='Probe_2_Sensor_Value')

rects5 = plt.bar(index + (4*(bar_width)), Probe_3_Sensor_Value, bar_width,alpha=opacity,color='k',label='Probe_3_Sensor_Value')

plt.ylabel('Turbidity Value')
plt.title('Turbidity Distribution for Every Standard Solution')
plt.xticks(index + bar_width, ('S_S_0','S_S_20','S_S_50','S_S_100','S_S_200','S_S_1000','S_S_4000'))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
n = 6

m1 = (0.10,0.12,0.10,0.11,0.14,0.10)
m2=(0.21,0.21,0.20,0.22,0.20,0.21)
m3=(0.29,0.27,0.28,0.24,0.23,0.23)
m4=(0.41,0.39,0.35,0.37,0.41,0.40)
x=[1,2,3,4,5,6]

fig, ax = plt.subplots()

index = np.arange(n)
bar_width = 0.2

opacity = 0.4
error_config = {'ecolor': '0.3'}
r1 = ax.bar(index, m1, bar_width,
                 alpha=opacity,
                 color='b',

                 error_kw=error_config)

r2 = ax.bar(index + bar_width, m2, bar_width,
                 alpha=opacity,
                 color='r',

                 error_kw=error_config)

r3 = ax.bar(index + bar_width+ bar_width, m3, bar_width,
                 alpha=opacity,
                 color='y',
                 error_kw=error_config)
r4 = ax.bar(index + bar_width+ bar_width+ bar_width, m4, bar_width,
                 alpha=opacity,
                 color='c',
                 error_kw=error_config)                 
plt.xlabel('D')
plt.ylabel('Anz')
plt.title('Th')

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')

ax1.bar(x,m1, 0.2) 
ax2.bar(x,m2, 0.2)
ax3.plot(x,m3)
ax4.plot(x,m4)

plt.tight_layout()
plt.show()
```
